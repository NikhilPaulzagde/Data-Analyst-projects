```python
# Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
sns.set_theme(style="whitegrid")  
pd.set_option("display.max_columns",100)
```

```python
# Outliers 

plt.figure(figsize=(10,10))
sns.boxplot(data= df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Numerical Column Analysis

plt.figure(figsize=(12, 8))


numerical_columns = ['Age', 'DailyRate', 'MonthlyIncome', 'PercentSalaryHike', 'TotalWorkingYears']

for i , col in enumerate(numerical_columns,1):
    plt.subplot(3,2,i)
    sns.histplot(df[col],kde= True, color='green', bins=20)
    plt.title(f"Distribution of {col}")
    
plt.tight_layout()
plt.show()
    
```

```python
categorical_column = df.select_dtypes(include='O').columns
categorical_column
```

```python
# Categorical Column Analysis

plt.figure(figsize=(12,15))



for i , col in enumerate(categorical_column,1):
    plt.subplot(3,3,i)
    sns.histplot(df[col],kde= True, color='green', bins=20)
    plt.xticks(rotation=90)
    plt.title(f"Distribution of {col}")
    
plt.tight_layout()
plt.show()
```

```python
# Correlation Analysis

correlation = df[numerical_columns].corr()

sns.heatmap(correlation, annot=True,cmap="Pastel1",cbar=False )
```

```python
# Age Distribution by Attrition

sns.histplot(data=df, x='Age', hue='Attrition', palette="Set2")
plt.title('Age Distribution by Attrition')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# JOb Leve Distribution by Attrition
 
sns.countplot(data=df, x='JobLevel', hue='Attrition', palette="Set2")
plt.title('JobLevel Distribution by Attrition')
plt.xlabel('Job Level')
plt.ylabel('Count')
plt.show()
```

```python
# JOb Role Distribution by Attrition
 
sns.countplot(data=df, x='JobRole', hue='Attrition', palette="Set2")
plt.title('Job Role Distribution by Attrition')
plt.xlabel('Job Role')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.show()
```

```python
# Year with Current Manager Distribution by Attrition
 
sns.countplot(data=df, x='YearsWithCurrManager', hue='Attrition', palette="Set2")
plt.title('Attrition by Years with Current Manager')
plt.xlabel('Years')
plt.ylabel('Count')
plt.show()
```

```python
# Education Distribution by Attrition
 
sns.countplot(data=df, x='EducationField', hue='Attrition', palette="Set1")
plt.title('Attrition by Education ')
plt.xlabel('Years')
plt.ylabel('Count')
plt.show()
```

```python
# Education Distribution by Attrition
 
sns.countplot(data=df, x='MaritalStatus', hue='Attrition', palette="Set1")
plt.title('Attrition by MaritalStatus')
plt.xlabel('Years')
plt.ylabel('Count')
plt.show()
```

```python
# Monthly Income Distribution

sns.histplot(data = df,x="MonthlyIncome" )
plt.title("Monthly Income Distribution")
plt.xlabel("Income")
plt.ylabel("Employees")
plt.show()
```

```python
# Salary Hike Distribution

sns.countplot(data = df,x="PercentSalaryHike" )
plt.title("Salary Hike Distribution")
plt.xlabel("Hike")
plt.ylabel("count")
plt.show()
```

```python
# Monthly Income vs Salary Hike

sns.scatterplot(data=df, x='MonthlyIncome', y='PercentSalaryHike', color='m')
plt.title('Monthly Income vs. Salary Hike')
plt.xlabel('Monthly Income')
plt.ylabel('Percent Salary Hike')
plt.show()
```

```python
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='MonthlyIncome', hue='Attrition', palette='Set2')
plt.title('Attrition by Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='PercentSalaryHike', hue='Attrition', palette='Set2')
plt.title('Attrition by Salary Hike')
plt.xlabel('Percent Salary Hike')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Tenure Distribution

sns.histplot(data=df, x='YearsAtCompany', kde=True, bins=20, color='green')
plt.title('Tenure (Years at Company) Distribution')
plt.xlabel('Years at Company')
plt.show()
```

```python
# Promotion

sns.countplot(data=df, x='YearsSinceLastPromotion', palette='Set1')
plt.title('Years Since Last Promotion Distribution')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Count')
plt.show()
```

```python
# Relationship between tenure and years since the last promotion

sns.scatterplot(data=df, x='YearsAtCompany', y='YearsSinceLastPromotion', color='blue')
plt.title('Tenure vs. Years Since Last Promotion')
plt.xlabel('Years at Company (Tenure)')
plt.ylabel('Years Since Last Promotion')
plt.show()
```
