```python
df['opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
```

```python
df.rename(columns={'coaster_name':'Coaster_Name',
                   'year_introduced':'Year_Introduced',
                 'opening_date_clean':'Opening_date',
                  'Inversions_clean':'Inversions',
                 'Gforce_clean':'Gforce' },inplace=True)
```

```python
df.columns
```

```python
df.duplicated(['Coaster_Name','Location']).sum()
```

```python
df=df.drop_duplicates('Coaster_Name')
```

```python
df=df.reset_index(drop=True).copy()
```

```python
df
```

```python
ax=df['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar' ,color='r', title='Top 10 Year Coaster Introduced')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
plt.show()
```

```python
ax=df['speed_mph'].plot(kind='hist' , title='Coaster Speed(mph)')
ax.set_xlabel('Speed(mph)')
ax.set_ylabel('Frequency')
```

```python
df.plot(kind='scatter',
        x='speed_mph' ,
        y='height_ft',
        color='r',
        title='Coaster Speed vs Height')
plt.show()
```

```python
ax=df.query('Location !="Other"')\
.groupby('Location')['speed_mph']\
.agg(['mean','count'])\
.query('count>=10')\
.sort_values('mean')['mean']\
.plot(kind='barh',figsize=(10,5),title='Average Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
